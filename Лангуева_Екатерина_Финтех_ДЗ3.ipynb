## Задание 2
Найдите доверительный интервал для оценки математического ожидания μ нормального распределения с уровнем значимости α = 0.1, зная
выборочное среднее X = 36.6, объем выборки N = 20 и среднее квадратическое отклонение σ = 2.2.

```python
from scipy.stats import norm
import numpy as np
```

```python
X = 36.6
N = 20
sigma2 = 2.2
alpha2 = 0.1
```

```python
z_alpha = norm.ppf(1 - alpha2 / 2)

error_margin = z_alpha * sigma2 / np.sqrt(N)

lower_bound = X - error_margin
upper_bound = X + error_margin
```

```python
print(f"Доверительный интервал для математического ожидания μ с уровнем значимости 0.1: ({lower_bound:.2f}, {upper_bound:.2f})")
```

# Задание 3

Известно, что в 2024 году средний балл за ЕГЭ по профильной математике составил 62.6 со стандартным отклонением 6.4. В некотором классе 20 учеников написали экзамен таким образом, что их средний результат оказался равным 66.1. На уровне статистической значимости α = 0.05 проверьте гипотезу H1 = ” Этот класс написал значительно лучше, чем в среднем по стране ”.

```python
mu_0 = 62.6
sigma3 = 6.4
X_bar = 66.1
N = 20
alpha3 = 0.05
```

```python
z_score = (X_bar - mu_0) / (sigma3 / np.sqrt(N))

z_critical = norm.ppf(1 - alpha3)
```

```python
print(f"Z-статистика: {z_score:.2f}")
print(f"Критическое значение Z для уровня значимости {alpha3}: {z_critical:.2f}")

if z_score > z_critical:
    print("Отвергаем гипотезу H0: Класс написал значительно лучше, чем в среднем по стране.")
else:
    print("Нельзя отвергнуть гипотезу H0: Недостаточно оснований утверждать, что класс написал лучше, чем в среднем по стране.")
```

# Задание 4

Поставщик кофе утверждает, что средний вес пачки кофе составляет ровно 100 г. Во время контрольной закупки из партии извлекается выборка из десяти пачек и взвешивается. Массы пачек (в граммах) равны: 97, 98, 104, 97, 101, 99, 99, 101, 98 и 100. Можно ли на уровне 99% доверять поставщику относительно средней массы пачки? Считайте, что вес пачки распределен нормально.

```python
from scipy.stats import t
```

```python
mu_04 = 100
sample = [97, 98, 104, 97, 101, 99, 99, 101, 98, 100]
alpha4 = 0.01
N = len(sample)
df = N - 1
```

```python
X_bar = np.mean(sample)
s = np.std(sample, ddof=1)

t_score = (X_bar - mu_04) / (s / np.sqrt(N))

t_critical = t.ppf(1 - alpha4 / 2, df)
```

```python
print(f"t-статистика: {t_score:.2f}")
print(f"Критическое значение t для уровня значимости {alpha4} и {df} степеней свободы: ±{t_critical:.2f}")

if abs(t_score) > t_critical:
    print("Отвергаем  гипотезу H0: Средний вес пачек статистически отличается от 100 г.")
else:
    print("Нельзя отвергнуть гипотезу H0: Нет оснований считать, что средний вес пачек отличается от 100 г.")
```

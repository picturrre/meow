## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
avg_df=ratings.groupby('movieId')['rating'].mean().reset_index()
final_df = pd.merge(
    movies, avg_df,
    left_on='movieId',
    right_on='movieId')
```

```python
for i in range(len(final_df)):
    if round(final_df.loc[i, 'rating']) <= 2:
        final_df.loc[i, 'class'] = 'низкий рейтинг'
    elif round(final_df.loc[i, 'rating']) <= 4:
        final_df.loc[i, 'class'] = 'средний рейтинг'
    elif round(final_df.loc[i, 'rating']) >= 4.5:
        final_df.loc[i, 'class'] = 'высокий рейтинг'  
        
final_df.sort_values('rating',ascending=False)
```

## Задание 2
Используйте файл keywords.csv.

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую при надлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток: 

**geo_data** = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

```python
import re
```

```python
keywords = pd.read_csv('keywords.csv')
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def classify_region(keyword):
    for region, cities in geo_data.items():
        for city in cities:
            if re.search(r'\b' + re.escape(city) + r'\b', keyword, re.IGNORECASE):
                return region
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(classify_region)
keywords.sort_values('region',ascending=True)
```

```python
avg_df2=keywords.groupby('region').size().reset_index(name='count').sort_values('count',ascending=True)
avg_df2
```
